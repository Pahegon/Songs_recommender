```python
def drop_columns(df_final):
    df_final.drop(columns=["Unnamed: 0.1","Unnamed: 0"], inplace=True)
    return df_final
```

```python
drop_columns(df_final)
```

```python
playlist = sp.user_playlist_tracks("spotify", "5lSgExb6yTKfLusGag7bm7",market="GB")
```

```python
def get_playlist_tracks(username, playlist_id):
    results = sp.user_playlist_tracks(username,playlist_id,market="GB")
    tracks = results['items']
    while results['next']:
        results = sp.next(results)
        tracks.extend(results['items'])
    return tracks
```

```python
tracks=get_playlist_tracks("spotify", "5lSgExb6yTKfLusGag7bm7")
```

```python
list_of_audio_features=[]
for item in range(0,1000):
    #print (tracks[item]["track"]["id"])
    list_of_audio_features.append(sp.audio_features(tracks[item]["track"]["id"])[0])
```

```python
df=pd.DataFrame(list_of_audio_features)    
df=df[["danceability","energy","loudness","speechiness","acousticness",
    "instrumentalness","liveness","valence","tempo","id","duration_ms"]]

df
```
