## Standardise the data using standardscaler

```python
import numpy as np
import pickle
from sklearn import datasets # sklearn comes with some toy datasets to practise
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from matplotlib import pyplot
from sklearn.metrics import silhouette_score
```

```python
#Check the infomation of the data points
df.describe()
```

```python
#We can see that the scale of sdt, mean... of loudness, tempo and duration_ms is much bigger than the other variables.
#We need to standarise them to the same level.
```

```python
#Before standardization we need to drop the no-numerical values
df.info()
```

```python
#Let's create a new dataframe (X) without no-numerical values ("id")
X = df[['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo','duration_ms']].copy()

X
```

```python
df
```

```python
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
X_scaled_df = pd.DataFrame(X_scaled, columns = X.columns)
display(X.head())
print()
display(X_scaled_df.head())
```

```python
X_scaled_df.describe()
```

```python
X_scaled_df
```

# Choosing the number of clusters k and  fitting k means cluster.

```python
#Initally I'm going to choose 4 clusters. Later will compare it with the Elbow method and see which one is better.

kmeans = KMeans(n_clusters=4, random_state=1234)
kmeans.fit(X_scaled_df)
```

```python
kmeans.labels_
```

```python
# assign a cluster to each example
labels = kmeans.predict(X_scaled_df)
# retrieve unique clusters
clusters = np.unique(labels)
# create scatter plot for samples from each cluster
for cluster in clusters:
    # get row indexes for samples with this cluster
    row_ix = np.where(labels == cluster)
    # create scatter of these samples
    pyplot.scatter(X.to_numpy()[row_ix, 1], X.to_numpy()[row_ix, 3])
    # show the plot
pyplot.show()
```

```python
#Now we have our "model", let's get the clusters and how many observations has each cluster.
clusters = kmeans.predict(X_scaled_df)
#clusters
pd.Series(clusters).value_counts().sort_index()
```

```python
#Now let's see to which group was assigned every song.

#X_df = pd.DataFrame(X)
X["cluster"] = clusters
X
```

# Build the elbow graph to find the best k.

```python
K = range(2, 21)
inertia = []

for k in K:
    print("Training a K-Means model with {} clusters! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_scaled_df)
    inertia.append(kmeans.inertia_)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
#As we can see, there is not appreciabel elbows in the chart. 
#I'm going to try with Silouhette

def load(filename = "filename.pickle"): 
    try: 
        with open(filename, "rb") as f: 
            return pickle.load(f) 
        
    except FileNotFoundError: 
        print("File not found!") 
```

```python
scaler2 = load("scaler.pkl")
```

```python
K = range(2, 20)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_scaled_df)
    
    filename = "scaler.pkl" + str(k) + ".pickle"
    with open(filename, "wb") as f:
        pickle.dump(kmeans,f)
    
    silhouette.append(silhouette_score(X_scaled_df, kmeans.predict(X_scaled_df)))


plt.figure(figsize=(16,8))
plt.plot(K, silhouette, 'bx-')
plt.xlabel('k')
plt.ylabel('silhouette score')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Silhouette Method showing the optimal k')
```

```python
# The best number of cluster is selected as the one with the best score.
# In this case I would say 8 clusters would be the right number of clusters for my Dataframe.
```
